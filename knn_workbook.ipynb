## Machine Learning Algorithm

In machine learning, there are three types:

1. **Supervised Learning**:  the model is trained on a labeled dataset, where each example is paired with the correct label.

2. **Unsupervised Learning**: In unsupervised learning, the model is given an unlabeled dataset and must find patterns and relationships

3. **Reinforcement Learning**: The model receives feedback in the form of rewards or penalties based on its actions and learns to maximize its reward over time.

### Supervised Learning

Under supervised learning, there are two main types:

1. **Classification**
2. **Regression**

train_test_split - splits a dataset into two subsets:

1). one for training a model

2). another for testing the model.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
df=pd.read_csv('diabetes.csv',index_col=0)
df.head()
```

Data Cleaning

```python
cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[cols] = df[cols].replace(0, np.nan)
df.head(10)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(),inplace=True)
df.head()
```

```python
df.isnull().sum()
```

Take x as independent variable

```python
X= df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness','Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']]
X
```

Take y as dependent Variable

```python
y=df[['Outcome']]
y
```

```python
y = y.replace({'No': 0, 'Yes': 1})
y
```

```python
df.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=42)
X_train
```

```python

X_test
```

```python
y_train
```

```python
y_test
```

KNN calling

trains the KNN classifier on the training data

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

Using trained KNN classifier to make predictions on the test data

```python
y_pred = knn.predict(X_test)
```

```python
 print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
 print("Accuracy:", accuracy_score(y_test, y_pred))
 #print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
for i in [5,7,9,11]:
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train,y_train)
  y_pred = knn.predict(X_test)
  print("ConfusionMatrix:\n",confusion_matrix(y_test,y_pred))
  print("Accuracy:",accuracy_score(y_test,y_pred))
#  print("ClassificationReport:\n",classification_report(y_test,y_pred))
```
